```python
emprestimos_data recebe
```

Explicando o codigo abaixo:

emprestimos_data é uma nova variável
está recebendo um DataFrame (pd.DataFrame) do DataFrame emprestimos_completo, na coluna 'data_emprestimo', onde esta sendo contado o numero de vezes que a data aparece(value_counts()), e sendo resetado o index (.reset_index())

emprestimo_data.columns =['data', 'quantidade']. reescreve o nome das colunas copiadas para essa nova DataFrame para 'data', 'quantidade'
pd.to_datetime(). configura a coluna data para data padrão do pandas (0000-00-00


```python
emprestimos_data = pd.DataFrame(emprestimos_completo['data_emprestimo'].value_counts()).reset_index()
emprestimos_data.columns = ['data','quantidade']
emprestimos_data['data'] = pd.to_datetime(emprestimos_data['data'])
emprestimos_data
```

Explicando o codigo abaixo:

variavel nova "emprestimos_por_ano" recebe variavel "emprestimos_data" 
emprestimos_data.data.dt.year - extrai o ano de cada data na coluna data
.groupby(by=) Agrupa o DataFrame pelos anos extraídos.
.sum() - Soma todas as colunas numéricas para cada grupo (ano).
['quantidade'].sum() - some a quantidade de cada ano que foi agrupado pelo .groupby(by=)

```python
emprestimos_por_ano = emprestimos_data.groupby(by=emprestimos_data.data.dt.year)['quantidade'].sum()
emprestimos_por_ano.index.name = 'ano'
emprestimos_por_ano
```
