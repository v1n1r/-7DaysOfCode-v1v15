Coletando dados com "pd.read_csv" 

O pd.read_csv() é uma das funções mais importantes e utilizadas na biblioteca pandas do Python. Ele serve para ler arquivos CSV (Comma-Separated Values) e transformá-los em um DataFrame, que é a estrutura de dados principal do pandas.

Funcionalidades principais:

- Importação de dados: Converte um arquivo CSV em um DataFrame pandas

- Leitura de URLs ou caminhos locais: Pode ler tanto arquivos locais quanto remotos (como do GitHub)

Tratamento automático:

- Reconhece cabeçalhos (headers)

- Infere tipos de dados das colunas

- Lida com valores faltantes

para importar dados do GitHub pode copiar o endereço onde o arquivo .csv está e adicionar "?raw=tue"

O que exatamente o "?raw=true" faz?

Comportamento do GitHub:

Sem ?raw=true:

GitHub mostra uma página HTML formatada

Contém interface gráfica, cabeçalhos, números de linha

Não é o conteúdo puro do CSV que o pandas precisa

Com ?raw=true:

GitHub retorna apenas o conteúdo bruto do arquivo

Remove todos os elementos da interface web

Parquet é um formato de arquivo colunar otimizado para armazenamento e processamento eficiente de dados estruturados em larga escala. Foi desenvolvido pelo Apache como parte do ecossistema Hadoop, mas tornou-se padrão no mundo de dados moderno.

"pd.read_parquet" é a função do Pandas para ler aquivos parquet

para preparar o link para a função ler corretamente os dados precisamos substituir "blob" por "raw" no link url onde os dados se encontram no GitHUb

ex:

url direto do GitHub: "https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/blob/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet"

ul com alteração de blob para raw "https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet"
